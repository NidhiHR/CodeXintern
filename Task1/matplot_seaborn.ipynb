```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('ToyotaCorolla.csv')


x=data['KM']
y=data['Doors']
z=data['Price']

ax=plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap='jet')
ax.set_title("3D PLOT")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('ToyotaCorolla.csv')

x=data['KM']
y=data['Weight']
z=data['Price']

plt.tricontourf(x,y,z,levels=20,cmap='jet')
plt.colorbar(label='Price')
plt.xlabel('KM')
plt.ylabel('Weight')
plt.title('cONTOUR PLT')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('ToyotaCorolla.csv')

sns.heatmap(data[['KM','Doors','Price','Weight','Age']].corr(),cmap='jet',annot=True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('ToyotaCorolla.csv')

plt.title('box plot')
plt.boxplot([data['KM'],data['Price'],data['Weight']])
plt.xticks([1,2,3],['KM','Price','Weight'])
plt.show()
```
